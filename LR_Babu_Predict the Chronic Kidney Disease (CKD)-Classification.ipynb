several parameters. The Client has provided the dataset of the same.

```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Reading the Dataset
dataset = pd.read_csv('CKD.csv')
```

```python
dataset
```

```python
#2.) Tell basic info about the dataset 
dataset.info()
```

```python
#(Total number of rows, columns)
dataset.shape
```

```python
#3.) Mention the pre-processing method if you’re doing any (like converting string to number – nominal data)
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
#converting string to number 
dataset=dataset.astype(int)
dataset
```

```python
#indep=dataset[['Age', 'EstimatedSalary','Gender_Male']]
#dep=dataset['Purchased']
```

```python
#x=dataset.drop(columns=['Purchased','User ID'])
x=dataset.drop(columns=['classification_yes'])
x
```

```python
y=dataset['classification_yes']
y
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#https://scikit-learn.org/stable/modules/model_evaluation.html#scoring-parameter
```

```python
#4.) Develop a good model with good evaluation metric. You can use any
##machine learning algorithm; you can create many models. Finally, you
#have to come up with final model.
```

```python
#LogisticRegression- its works as Classsification
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

param_grid = {'solver':['newton-cg', 'lbfgs', 'liblinear', 'saga'],
             'penalty':['l2']} 

#solver -optimize algorithm used to find the optimal weights for the classifier.
#penalty:regularization technique used to preventr overfitting. lasso(l1 reg),ridge reg(l2), elasticnet(l1+l2),nono:no reg

grid = GridSearchCV(LogisticRegression(), param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1_weighted') 
#verbos : control logging o/p during training  
#most use cases, verbose=3 is sufficient to monitor the training process
# fitting the model create for grid search 
grid.fit(X_train, y_train) 


```

```python
#verbose=0: No output (silent).
#verbose=1: Minimal output (e.g., general progress).
#verbose=2: Moderate output (e.g., updates per epoch or batch).
#verbose=3: Detailed output.
#verbose=4: Very detailed, debug-level output (if supported).
```

```python
# print best parameter after tuning 
#print(grid.best_params_) 
re=grid.cv_results_
#print(re)  table format
grid_predictions = grid.predict(X_test) 
#y_pred   

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, grid_predictions)



# print classification report 
from sklearn.metrics import classification_report
clf_report = classification_report(y_test, grid_predictions)


```

```python
print("The confusion Matrix:\n",cm)
```

```python
print("The report:\n",clf_report)
```

```python
#best parameters
from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,grid_predictions,average='weighted')
print("The f1_macro value for best parameter {}:".format(grid.best_params_),f1_macro)
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(y_test,grid.predict_proba(X_test)[:,1])
#probability estimates for each class for the input data.(such as logistic regression, random forests, and gradient boosting.)
#Column 0: Predicted probabilities of class 0 (negative class).
#Column 1: Predicted probabilities of class 1 (positive class).
#[:,1] --predicted probabilities for the positive class (y_pred_proba), not the class labels (y_pred).
```

```python
#Receive operator curve-ROC curve:is a graphical representation of a classifier's performance across different threshold values.
#True Positive Rate (TPR) (Sensitivity or Recall) on the y-axis.#False Positive Rate (FPR) (1 - Specificity) on the x-axis

#Area Under Curve-AUC:is a single scalar value that measures the area under the ROC curve. 
#It provides a summary of the model's ability to distinguish between classes
#AUC = 1: Perfect model (ideal separation of classes).
#AUC = 0.5: Random guessing (no discriminatory power).
#AUC < 0.5: Worse than random (indicates potential issues).
```
